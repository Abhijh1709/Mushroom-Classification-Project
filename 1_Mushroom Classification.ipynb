#### Data Manipulation

```python
data.info()
```

```python
data.astype('category')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in data.columns:
    data[column]= le.fit_transform(data[column])
```

```python
data.head()
```

#### Store Feature Matrix in X and Response (target) in vector y

```python
x = data.drop('class', axis=1)
y = data['class']
```

```python
x
```

```python
y
```

```python
from sklearn.decomposition import PCA
pca1 = PCA(n_components=7)
pca_fit = pca1.fit_transform(x)
```

```python
pca1.explained_variance_ratio_
```

```python
sum(pca1.explained_variance_ratio_)
```

#### splitting the dataset into the training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(pca_fit,y,test_size=0.2, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

#### Import the models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

#### Model Training

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)

knn = KNeighborsClassifier()
knn.fit(x_train,y_train)

svc = SVC()
svc.fit(x_train,y_train)

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)

rm = RandomForestClassifier()
rm.fit(x_train,y_train)

gb = GradientBoostingClassifier()
gb.fit(x_train,y_train)
```

### Prediction on Test Data

```python
y_pred1 = lr.predict(x_test)
y_pred2 = knn.predict(x_test)
y_pred3 = svc.predict(x_test)
y_pred4 = dt.predict(x_test)
y_pred5 = rm.predict(x_test)
y_pred6 = gb.predict(x_test)
```

#### Evaluating the Algotithm

```python
from sklearn.metrics import accuracy_score
```

```python
print("ACC LR = ", accuracy_score(y_test, y_pred1))
print("ACC KNN =", accuracy_score(y_test, y_pred1))
print("ACC SVC =", accuracy_score(y_test, y_pred1))
print("ACC DT =", accuracy_score(y_test, y_pred1))
print("ACC RM =", accuracy_score(y_test, y_pred1))
print("ACC GBC =", accuracy_score(y_test, y_pred1))
```

```python
final_data = pd.DataFrame({'Models':['LR','KNN','SVC','DT','RM','GBC'],
                          'ACC': [accuracy_score(y_test, y_pred1)*100,
                                 accuracy_score(y_test, y_pred2)*100,
                                 accuracy_score(y_test, y_pred3)*100,
                                 accuracy_score(y_test, y_pred4)*100,
                                 accuracy_score(y_test, y_pred5)*100,
                                 accuracy_score(y_test, y_pred6)*100]})
```
